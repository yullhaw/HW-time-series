Создайте столбец source_type для датафрейма log из материалов занятия по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic;
для источников paid и email из России — ставится ad;
для источников paid и email не из России — ставится other;
все остальные варианты берутся из traffic_source без изменений.

```python
import pandas as pd
```

```python
log=pd.read_csv('visit_log_.csv', sep=';')
log.head()
```

```python
def filtered(row):
    if row['traffic_source']=='yandex':
        return 'organic'
    elif row['traffic_source']=='google':
        return 'organic'
    elif row['traffic_source']=='paid':
        if row['region']=='Russia':
            return 'ad'
        else:
            return 'other'
    elif row['traffic_source']=='email':
        if row['region']=='Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return row['traffic_source']
```

```python
log['source_type']=log.apply(filtered, axis=1)
```

Откройте файл ratings.csv с оценками фильмов.
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
Примечание: под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
import numpy as np
tmp = pd.read_csv('ml-latest-small/ratings.csv')
tmp
```

```python
uid_count_lt = tmp.groupby('userId')['timestamp']\
        .agg({'count',  lambda x: max(x) - min(x)}).reset_index()\
        .rename(columns={'<lambda_0>':'lt'})
# mean lt for users, wich make >100 ratings
uid_count_lt[uid_count_lt['count'] > 100]['lt'].mean()
```
